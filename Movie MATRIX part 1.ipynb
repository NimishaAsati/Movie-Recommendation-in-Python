```python
##Q2.a

# Creating the Xij matrix which gives the values, how many users liked both movies i and j through each of its value
movieIDs = np.unique(np.array(ratings_data["movieId"]))

## We are considering if the movie ratings are 3 , 3.5 , 4, 4.5 ,5 , users are liking the movies.
liked_movies = ratings_data[ratings_data.rating > 2.5]                               
X_mat= []
X_mat = pd.DataFrame(0,index = movieIDs, columns = movieIDs)         

## I is the each user here , User group is the corresponsing all movies and ratings info records corresponding to each user
for i, User_group in liked_movies.groupby(['userId']):
    movie_list = list(User_group.movieId)
    for i in range(len(movie_list)):
        for j in range(i+1,len(movie_list)):
            X_mat.ix[movie_list[i],movie_list[j]] += 1
            X_mat.ix[movie_list[j],movie_list[i]] += 1  

```

```python
## To print and check if i reached the last user and view the Xij values 
print(i) # 610
Xij = np.array(X_mat)
Xij_len = Xij.shape[1]
##Xij_len #9724
X_mat
#X_mat[0] #9724
#X_mat[1] #9724
```

```python
vect_dim = 300 
vec_mat = np.random.uniform(-1,1,size=(vect_dim,Xij_len)) ### INTERCHANGE
vec_mat
```

```python
##Q2.b

## Gradient Descent
reg = 1e-1
stepsize = 5e-3
loss_df = []

for i in range(1500):
    
    # Calculating the scores
    cost_mat = (np.dot(vec_mat.T,vec_mat) - Xij)
    
    for m in range(cost_mat.shape[1]):
        cost_mat[m,m] = 0 
    
    #Calculate loss
    sq_loss = np.square(cost_mat)
    data_loss = np.sum(sq_loss)/Xij_len
    reg_loss = 0.5*reg*np.sum(vec_mat*vec_mat) #using L2 regularization
    cost = data_loss + reg_loss
    #print(cost)

    loss_df.append(cost)
    
    
    dV = ((4 * np.dot(cost_mat,vec_mat.T))/Xij_len) + reg*vec_mat.T
    
    vec_mat -= stepsize * dV.T
    if i % 100 == 0:
            print ("iteration %d: loss %f" % (i, cost))
            
```
